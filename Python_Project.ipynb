```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('DS1_C8_V3_ND_Sprint2_Data Analysis Using Python_Dataset.csv')
df
```

# Task 1

```python
#Display number of rows and columns in data set
df.shape
```

```python
#Display genre and title of the first 50 movies
df[['genres','title']].head(50)
```

# Task 2

```python
df.isnull().sum()
```

```python
#Replacing Null Values
df.homepage=df.homepage.fillna('Not Defined')
df.overview=df.overview.fillna('Not Defined')
df.tagline=df.tagline.fillna('Not Defined')
df.release_date=df.release_date.fillna(df.release_date.mode()[0])
df.runtime=df.runtime.fillna(df.runtime.median())
```

```python
df.isnull().sum()
```

# Task 3

```python
#Movie category that have budget more then $220,000
cond=df['budget']>220000
t3=df[cond]
print(t3['genres'])
print('genere:',t3['genres'].size)
```

# Task 4

```python
#Movie category that have revenue more then $961,000,000
cond=df['revenue']>961000000
t4=df[cond]
print(t4['genres'])
print('genere:',t4['genres'].size)
```

# Task 5

```python
#Droping the rows where Value is zero for both revenue and budget 
cond=(df['budget']==0)|(df['revenue']==0)
t5=df[cond]
df.drop(index=t5.index,inplace=True)
shape=df.shape
print(f'Rows after droping: {shape[0]}')
```

# Task 6

```python
#Top 10 movie with the highest revenue and top 10 movie with the least budget
t6_1=df[['title','revenue']].sort_values('revenue',ascending=False).head(10)
t6_2=df[['title','budget']].sort_values('budget',ascending=True).head(10)
fig,axs=plt.subplots(2,1,figsize=(15,10))
sns.barplot(data=t6_1,x='title',y='revenue',ax=axs[0])
axs[0].set_title('Top 10 movie with the highest revenue')
sns.barplot(data=t6_2,x='title',y='budget',ax=axs[1])
axs[1].set_title('Top 10 movie with the least budget')
```
